```python
# 필요한 모듈 불러오기
import pandas as pd

df = pd.read_csv("C:/Users/tjoeun/Desktop/mm/xuyite/csv/preprocess.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm

dfX_train, dfX_test, dfy_train, dfy_test = train_test_split(dfX, dfy, test_size=0.2, random_state=10)

model = sm.OLS(dfy_train, dfX_train)
result = model.fit()
result.summary()
```

## 다중공선성

```python
dfX.corr()
```

```python
# 'FG', 'FGA', '3P', '3PA', '2P', '2PA','FT', 'FTA', 'TRB' 제거
dfX.drop(['FG', 'FGA', '3P', '3PA', '2P', '2PA','FT', 'FTA', 'TRB'], axis=1, inplace=True)
```

```python
# 다중공선성 테스트
dfX_train, dfX_test, dfy_train, dfy_test = train_test_split(dfX, dfy, test_size=0.2, random_state=10)

model = sm.OLS(dfy_train, dfX_train)
result = model.fit()
result.summary()
```

```python
# 후진제거법
dfX.drop(['3P%'],axis=1,inplace= True)
dfX.drop(['FT%'],axis=1,inplace= True)
```

```python
# 데이터 합치기
df1 = pd.concat([dfX, dfy], axis=1)
df1
```

```python
# 컬럼명 변경
df1.rename(columns = {'salary':'inflationAdjSalary'}, inplace = True)
```

```python
# 연봉과의 관계도
# 제일 연봉에 가장 상관을 미칠만한 PTS(총득점수),DRB(수비형 리바운드),GS(주전으로 시작한 경기수) 3가지를 분석
df1.corr()['inflationAdjSalary'].sort_values()
```

```python
# 연봉과 가장 높은 상관관계를 지닌 3가지 변수의 연봉과의 관계도
# 연봉과 PTS FG FGA 관계(pairplot)
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(df1[["inflationAdjSalary", "PTS", "DRB", "GS"]])
plt.show()
# 양의 관계를 보이는것으로 보임
```

```python
# 상관계수의 검점
# PTS ,FG, FGA,과 연봉의 상관관계
Y = df1.inflationAdjSalary.values
from scipy import stats
import numpy as np
for item in ['PTS','GS','DRB']:
    print(item)
    X = df1[item].values
    print ('Covariancce : {:.2f}'.format(np.cov(X,Y)[0,1]))
    print ('Correlation : {:.2f}'.format(stats.pearsonr(X,Y)[0]))
    print ('P-value : {:.4f}'.format(stats.pearsonr(X,Y)[1]))
    print('\n') 
```
